```python
import csv
import random

def read_and_write(input_file, output_file):
    categories = ["Продукты", "Электроника", "Товары для животных","Бытовая техника", "Одежда", "Книги", "Товары для детей"]
#cчитываем файл
    with open(input_file, 'r', encoding='utf-8') as f_input, open(output_file, 'w', encoding='utf-8', newline='') as f_output:
        csv_reader = csv.reader(f_input, delimiter='\t')
        csv_writer = csv.writer(f_output)
        csv_writer.writerow(['user_id ', ' source ', ' category '])
        next(csv_reader)
# пройдемся по каждому юзеру 
        for users in csv_reader:
#извлеченм данные из строки, разделив ее (а мы знаем,в каком формате таблциа), присвоим значение user_id и source
            user_id, source = users[0].split(",")
            if user_id == "0":
                continue
            category = random.choice(categories)
            csv_writer.writerow([user_id, source, category])
```

```python
read_and_write('visit_log__1_.csv', 'funnel.csv')

with open('funnel.csv', 'r', encoding='utf-8') as result:
    for _ in range(150):
        print(result.readline().strip())
```
